В проекте используются данные о недвижимости в Бостоне за 7 июля 1993 года. Данные представлены в 10 столбцах:
1. CRIM (Уровень преступности на душу населения)
2. ZN (доля жилых земель, зонированных на участки более 25000 кв. Футов)
3. INDUS (доля площадей под неторговую деятельность на город)
4. CHAS Charles River (Фиктивная переменная (= 1, если река; 0 иначе))
5. NOX (Концентрация оксидов азота)
6. RM (Среднее количество комнат в жилом помещении)
7. AGE (Доля квартир, занимаемых владельцами, построенных до 1940 г.)
8. DIS (расстояния до пяти центров занятости Бостона
)
9. RAD (Индекс доступности радиальных магистралей)
10. TAX (Ставка налога на имущество на полную стоимость за 10 000 долларов США)
11. PTRATIO (Соотношение учеников и учителей)
12. B (доля черного населения в городе)
13. LSTAT (более низкий статус населения)
14. MEDV (Средняя стоимость домов, занимаемых владельцами, в 1000 долларов США)

*Результаты проекта могут использоваться в качестве основы для разработки модели машинного обучения для анализа цен для недвижимости.*


## II. Датасет
  *Датасет выгружен по ссылке https://www.kaggle.com/datasets/arslanali4343/real-estate-dataset/data*


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import CSV file as DataFrame
estate = pd.read_csv('data.csv')

#view DataFrame
estate
```

```python
estate.columns #смотрим названия столбцов
```

## III. Анализ и визуализация данных

```python
estate.hist(bins=50, figsize=(20, 15))
```

```python
estate['CRIM'].plot(kind = 'hist')
plt.xlabel('Уровень криминала по возрастанию')
plt.ylabel('Количество домов')
plt.title('Криминал')
plt.show()
```

```python
plt.hist(estate['NOX'], bins=20, color='skyblue', edgecolor='black') #строим гистограмму концентрации азота по возрастанию
```

Смотрим корреляции

```python
corr_matrix = estate.corr()
corr_matrix['MEDV'].sort_values(ascending=False)
```

```python
from pandas.plotting import scatter_matrix
attributes = ["MEDV", "RM", "ZN", "LSTAT"]
scatter_matrix(estate[attributes], figsize = (12,8))
```

```python
estate.plot(kind="scatter", x="RM", y="MEDV", alpha=0.8)
```

Совмещаем атрибуты и смотрим корреляции

```python
estate["TAXRM"] = estate['TAX']/estate['RM']
```

```python
estate.head()
```

```python
corr_matrix = estate.corr()
corr_matrix['MEDV'].sort_values(ascending=False)
```

```python
estate.plot(kind="scatter", x="TAXRM", y="MEDV", alpha=0.8)
```

Визуализируем важное, с использованием (EDA) анализа основных свойств данных, нахождения в них общих закономерностей

```python
fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(22,18))
fig.subplots_adjust(hspace=0.25)

for ax, col in zip(axes.flatten(), estate.columns):
    ax.hist(estate[[col]].values, bins=100)
    ax.set(title=str(col).lower() + " distribution")
```

```python
s_fig, s_ax = plt.subplots(figsize=(16, 16))

cols = estate.columns
sns.heatmap(estate[cols].corr(), ax=s_ax, cmap='RdBu_r', annot=True, center=0.0)
```
